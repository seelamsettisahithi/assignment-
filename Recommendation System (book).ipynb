```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv("C://data science//data sets//book.csv", encoding = "latin1")
```

```python
book.head()
```

```python
book.columns = [ "book_id","user_id","book_title","book_rating"]
```

```python
book[0:5]
```

```python
#number of unique users in the dataset
len(book.user_id.unique())
```

```python
len(book.book_id.unique())
```

```python
user_book_df = book.pivot(index='user_id',
                                 columns='book_id',
                                 values='book_rating').reset_index(drop=True)
```

```python
user_book_df.index = book.user_id.unique()
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.user_id.unique()
user_sim_df.columns = book.user_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
