```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import spacy
from bs4 import BeautifulSoup as bs 

import requests
```

```python
cust_names=[]
rating=[]
review_title=[]
reviews=[]
review_len=[]
cou=0
for i in range(1,299):
  url='https://www.amazon.in/Canon-1500D-Digital-Camera-S18-55/product-reviews/B07BS4TJ43/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber='+str(i)

  # headerAgent = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'}
  headerAgent = {
'authority': 'www.amazon.com',
'pragma': 'no-cache',
'cache-control': 'no-cache',
'dnt': '1',
'upgrade-insecure-requests': '1',
'user-agent': 'Mozilla/5.0 (X11; CrOS x86_64 8172.45.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.64 Safari/537.36',
'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
'sec-fetch-site': 'none',
'sec-fetch-mode': 'navigate',
'sec-fetch-dest': 'document',
'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
}
  page=requests.get(url,headers=headerAgent)
  # print(page)
  
  soup=bs(page.content,'html.parser')

  name= soup.find_all('span',class_='a-profile-name')
  name.pop(0)
  name.pop(0)
  title= soup.find_all('a',class_='review-title-content')
  rate = soup.find_all('i', {"data-hook":"review-star-rating"})
  rev = soup.find_all('span',{'data-hook':'review-body'})

  for j in range(0,len(name)):
     cust_names.append(name[j].text)

  for j in range(len(rev)):
    rating.append(float(rate[j].text[:3]))

    review_title.append(title[j].get_text().strip('\n'))
    reviews.append(rev[j].text.strip('  \n'))
    review_len.append(len(rev[j].text.strip('  \n')))

  # print(i)

# print(cust_names)

print(rating)
print(review_title)
print(reviews)
# print(cou)
```
